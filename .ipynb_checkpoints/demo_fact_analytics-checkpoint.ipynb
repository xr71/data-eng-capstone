```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.enableHiveSupport().getOrCreate()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fct_df = spark.read.parquet("./output_data/fct_immigration/")
```

```python
# fct_df.show(2)
```

```python
age_counts = fct_df.groupBy("age").sum("i94_counts").toPandas()
```

```python
plt.figure(figsize=(16,10))
plt.bar(age_counts["age"], age_counts["sum(i94_counts)"])
```
